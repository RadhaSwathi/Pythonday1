```python
Titanic_c=pd.concat([Titanic_train,Titanic_test],axis=0,sort=False,ignore_index=True)
Titanic_c.head()
```

```python
Titanic_c.isnull().sum()
```

```python
Titanic_c["Embarked"].fillna(value='S',inplace=True)
```

```python
Titanic_c[Titanic_c["Fare"].isnull()]
```

```python
Titanic_c[Titanic_c["Pclass"]==3]
```

```python
Titanic_c['Fare'].fillna(value=Titanic_c.groupby('Pclass')['Fare'].median()[3],inplace=True)
```

```python
[Titanic_c['Name'][i].split('.')[0].split(',')[1] for i in range(len(Titanic_c))]
```

```python
Title=pd.Series([Titanic_c['Name'][i].split(',')[1].split('.')[0].split(',')[0].strip() for i in range(len(Titanic_c))])
Titanic_cp=pd.concat((Titanic_c,Title.rename('Title')),axis=1,sort=False,)
Titanic_cp.head()
```

```python
Titanic_cp.groupby(Title)['Age'].median()['Mrs']
#pd.crosstab(Titanic_cp['Title'],Titanic_cp['Age'])
```

```python
Titanic_cp[(Titanic_cp["Age"].isna()==True) & (Titanic_cp['Title']=='Mrs')]['Age']=Titanic_cp[(Titanic_cp["Age"].isna()==True) & (Titanic_cp['Title']=='Mrs')]['Age'].fillna(value=Titanic_cp.groupby(Title)['Age'].median()['Mrs'],inplace=True)
```
